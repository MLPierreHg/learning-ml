```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

# 1) Import du jeu de données "iris"

```python
iris = load_iris()

x = iris.data
y = iris.target

names = list(iris.target_names)

print(f"x contient {x.shape[0]} exemples et {x.shape[1]} variables")
print(f"il y a {np.unique(y).size} classes")
```

```python
names
```

### Construire une fonction permettant de construire un graphique mettant en avant les différentes variables d'un dataset

```python
def multi_graphs(data,target):
    n = data.shape[1]
    plt.figure(figsize=(12,8))
    for i in range(n):
        # Le double slash (//) permet de s'assurer que la valeur que l'on obtient est bien une valeur entière.
        plt.subplot(n//2, n//2,i+1)
        plt.scatter(data[:,0], data[:,i],c=target)
        plt.xlabel('')
        plt.ylabel(i)
        plt.colorbar(ticks=list(np.unique(target)))
    plt.show
```

```python
multi_graphs(x,y)
```

```python
f = lambda x, y: np.sin(x) + np.cos(x+y)
f_two = lambda x, y: np.sin(x) + np.cos(x+y)*np.cos(x)
```

```python
X = np.linspace(0,5,100)
Y = np.linspace(0,5,100)

X, Y = np.meshgrid(X,Y)

Z = f(X,Y)
Z_two = f_two(X,Y)
```

```python
plt.figure(figsize=(15,15))
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, cmap='plasma')
```
