```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,2,10)
y = x**2
```

```python
print(x)
```

# Pyplot

```python
# couleur de la courbe
plt.plot(x,y,c='red')
```

```python
# taille du tracé
plt.plot(x,y,lw=10)
```

```python
# Style du trait (en petit tiret)
plt.plot(x,y,ls='--')
```

```python
plt.scatter(x,y)
```

```python
# Cycle de vie d'une figure
plt.figure(figsize=(12,8)) #fenêtre de travail pour la figure
plt.plot(x,y, label="quadratique")
plt.plot(x,x**3, label='cubique')
plt.title('Figure 1')
plt.xlabel('axe x')
plt.ylabel('axe y')
plt.legend()
plt.show()
plt.savefig('figure.png')

plt.figure()
plt.plot(x,y)
```

```python
plt.subplot(2,1,1)
plt.plot(x,y,c='red')
plt.title('courbe1')
plt.subplot(2,1,2)
plt.plot(x,x**3,c='blue')
plt.title('courbe2')
```

```python
plt.figure()
plt.subplot(2,2,1)
plt.plot(x,y,c='red')
plt.title('Graphique 1')
plt.subplot(2,2,2)
plt.plot(x,np.sin(x),c='blue',label='sinus')
plt.plot(x,np.cos(x),c='cyan', label='cosinus')
plt.title('Graphique 2')
plt.legend()
```

# Methode Orientée Objet

```python
fig, ax = plt.subplots(4,1,sharex=True)
ax[0].plot(x,y)
ax[1].plot(x,np.sin(x))
ax[2].plot(x,np.cos(x))
ax[3].plot(x,np.tan(x))
plt.show()
```

```python
print('Type de ax: ' + str(type(ax)) + ', taille de ax: ' + str(ax.shape))
```

# Exercice

```python
dataset = {f"experience{i}": np.random.randn(100,3) for i in range(5)}
```

```python
len(dataset)
```

```python
def graphique(dataset):
    fig, ax = plt.subplots(len(dataset),1,sharex=True)
    for i in range(0,len(dataset)):
        data = dataset.get(f"experience{i}")
        ax[i].plot(data)
    plt.show()
```

```python
print(graphique(dataset))
```

# Correction

```python
def graphique(data):
    n = len(data)
    plt.figure(figsize=(12,18))
    for k, i in zip(data.keys(),range(1,n+1)):
        plt.subplot(n,1,i)
        plt.plot(data[k])
        plt.title(k)
    plt.show()
```

```python
graphique(dataset)
```
